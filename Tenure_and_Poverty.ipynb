```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
pd.set_option('display.max_rows', 500)
```

```python
df = pd.read_csv('Tenure Type of the Household-final.csv')
```

```python
df.info()
```

```python
# Replacing column names.
df.columns = [
    "Financial year",
    "Median net household income(BHC)",
    "Median net household income(AHC)",
    "Housing benefit",
    "Tenure type",
    "Region",
    "Total population",
    "Extra1",
    "Extra2",
    "Extra3"
]
df = df.drop(columns=["Extra1", "Extra2", "Extra3"], axis=1, errors="ignore")

# Dropping the first row.
df = df.drop(index=0)

# Resetting the index.
df = df.reset_index(drop=True)
```

```python
# Filling missing values in the 'Financial year', 
# 'Median net household income(BHC)', 'Median net household income(AHC)', 
# 'Housing benefit','Tenure type'.
for col in enumerate(df.columns.to_list()[0:-2]):
    df[col[1]] = df[col[1]].ffill()

df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Converting 'Total population' column to numeric type.
df['Total population'] = pd.to_numeric(df['Total population'], errors='coerce')
df.head()
```

```python
df.shape
```

```python
# Checking for duplicates.
duplicates = df[df.duplicated()]
duplicates
```

```python
df.isnull().sum()
```

```python
df['Financial year'].unique().tolist()
```

```python
# Cleaning 'Financial Year' column.
df['Financial year'] = df['Financial year'].map({
    '2002-03 (cpi, r)' : '2002-03',
    '2003-04 (cpi, r)' : '2003-04',
    '2004-05 (cpi, r)' : '2004-05',
    '2005-06 (cpi, r)' : '2005-06',
    '2006-07 (cpi, r)' : '2006-07',
    '2007-08 (cpi, r)' : '2007-08',
    '2008-09 (cpi, r)' : '2008-09',
    '2009-10 (cpi, r)' : '2009-10',
    '2010-11 (cpi, r)' : '2010-11',
    '2011-12 (cpi, r)' : '2011-12',
    '2012-13 (cpi, r)' : '2012-13',
    '2013-14 (cpi, r)' : '2013-14',
    '2014-15 (r)' : '2014-15',
    '2015-16 (r)' : '2015-16',
    '2016-17 (r)' : '2016-17',
    '2017-18 (r)' : '2017-18',
    '2018-19 (r)' : '2018-19',
    '2019-20' : '2019-20',
    '2020-21 (covid2021)' : '2020-21',
    '2021-22 (covid2122, div)' : '2021-22',
    '2022-23 (colp, covid2223, div)' : '2022-23'
})

df.head()
```

```python
# Removing data in '2020-21' due to data quality concerns.
df =  df[df['Financial year'] != '2020-21']
```

```python
# Filling missing values with zero because there are no data available.
df = df.fillna(0)
```

```python
df.isnull().sum()
```

```python
# Checking unique values in the 'Region' column.
df['Region'].unique()
```

```python
# Cleaning 'Region' column data.
df['Region'] = df['Region'].map({
    'North East (E12000001)' : 'North East',
    'North West (E12000002)' : 'North West',
    'Yorkshire and The Humber (E12000003)' : 'Yorkshire and The Humber',
    'East Midlands (E12000004)' : 'East Midlands',
    'West Midlands (E12000005)' : 'West Midlands',
    'East (E12000006)' : 'East',
    'London (E12000007)' : 'London',
    'South East (E12000008)' : 'South East', 
    'South West (E12000009)' : 'South West',
    'Wales (W92000004)' : 'Wales',
    'Scotland (S92000003)' : 'Scotland',
    'Northern Ireland (N92000002)' : 'Northern Ireland'
})
```

```python
df.head()
```

```python
# Checking unique values in the 'Median net household income(BHC)' column.
df['Median net household income(BHC)'].unique().tolist()
```

```python
# Cleaning 'Median net household income(BHC)' column data.
df = df.copy()

df['Median net household income(BHC)'] = df[
    'Median net household income(BHC)'
].map({
    'Not in low income (at or above threshold)' : 'Not in low income',
    'In low income (below threshold)' : 'In low income'
})

df.head()
```

```python
# Checking unique values in the 'Median net household income(AHC)' column.
df['Median net household income(AHC)'].unique().tolist()
```

```python
# Cleaning 'Median net household income(AHC)' column data.
df['Median net household income(AHC)'] = df[
    'Median net household income(AHC)'
].map({
    'Not in low income (at or above threshold)' : 'Not in low income',
    'In low income (below threshold)' : 'In low income'
})

df.head()
```

```python
# Checking unique values in the 'Housing benefit' column.
df['Housing benefit'].unique()
```

```python
# Checking unique values in the 'Tenure type' column.
df['Tenure type'].unique()
```

```python
# Exploring Population Count Over Time by Housing benefit.
housing_benefit_over_time = df.groupby(['Financial year','Housing benefit'])\
    ['Total population'].sum()
housing_benefit_over_time
```

```python
housing_benefit_over_time.unstack().plot(kind='line', figsize=(14, 7))
plt.title('Population Count Over Time by Housing Benefit')
plt.xlabel('Financial Year')
plt.ylabel('Total Population')
plt.legend(title='Housing Benefit')
plt.grid(True)
plt.show()
```

```python
# Exploring Population Count Over Time by Tenure type.
tenure_type_over_time = df.groupby(['Financial year', 'Tenure type'])\
    ['Total population'].sum()
tenure_type_over_time
```

```python
tenure_type_over_time.unstack().plot(kind='line', figsize=(14, 7))
plt.title('Population Count Over Time by Tenure Type')
plt.xlabel('Financial Year')
plt.ylabel('Total Population')
plt.legend(title='Tenure Type')
plt.grid(True)
plt.show()
```

```python
# Creating a pivot table to find the relationship between BHC and housing benefit
bhc_housing_benefit = df.pivot_table(
    index='Median net household income(BHC)',
    columns='Housing benefit',
    values='Total population',
    aggfunc='sum'
)

# Plotting the pivot table
bhc_housing_benefit.plot(kind='bar', figsize=(14, 7))
plt.title('Relationship between BHC and Housing Benefit')
plt.xlabel('Median net household income(BHC)')
plt.ylabel('Total Population')
plt.legend(title='Housing Benefit')
plt.grid(True)
plt.show()
```
