```python
# Exploring Tenure Type distribution.
plt.figure(figsize=(10,6))
tenure_dist = df.groupby('Tenure type')['Total population'].sum().\
    sort_values(ascending=False)
tenure_dist.plot(kind='bar', color='teal')
plt.title('Population Distribution by Tenure Type')
plt.ylabel('Total Population (Millions)')
plt.gca().yaxis.set_major_formatter\
    (plt.FuncFormatter(lambda x, _: f'{x/1e6:.1f}M'))
plt.show()
```

```python
# Exploring Housing benefit distribution.
plt.figure(figsize=(10,6))
housing_benefit_dist = df.groupby('Housing benefit')['Total population'].sum().\
    sort_values(ascending=False)
housing_benefit_dist.plot(kind='bar', color='teal')
plt.title('Population Distribution by Housing Benefit')
plt.ylabel('Total Population (Millions)')
plt.gca().yaxis.set_major_formatter\
    (plt.FuncFormatter(lambda x, _: f'{x/1e6:.1f}M'))
plt.show()
```

```python
# Exploring Median net household income Before Housing Cost distribution.
plt.figure(figsize=(10,6))
tenure_dist = df.groupby('Median net household income(BHC)')['Total population'].sum().\
    sort_values(ascending=False)
tenure_dist.plot(kind='bar', color='purple')
plt.title('Population Distribution by Median net household income(BHC)')
plt.ylabel('Total Population (Millions)')
plt.gca().yaxis.set_major_formatter\
    (plt.FuncFormatter(lambda x, _: f'{x/1e6:.1f}M'))
plt.show()
```

```python
# Exploring Median net household income After Housing Cost distribution.
plt.figure(figsize=(10,6))
tenure_dist = df.groupby('Median net household income(AHC)')['Total population'].sum().\
    sort_values(ascending=False)
tenure_dist.plot(kind='bar', color='teal')
plt.title('Population Distribution by Median net household income(AHC)')
plt.ylabel('Total Population (Millions)')
plt.gca().yaxis.set_major_formatter\
    (plt.FuncFormatter(lambda x, _: f'{x/1e6:.1f}M'))
plt.show()
```

```python
# Exploring how Median net household income Before Housing Cost and 
# After Housing Cost are changing over the years for the In low income 
# gruop.

# Filtering the dataframe for 'In low income' in both 
# 'Median net household income(BHC)' and 'Median net household 
# income(AHC)'.
bhc_income_data = df[df['Median net household income(BHC)'] == 'In low income']
ahc_income_data = df[df['Median net household income(AHC)'] == 'In low income']

# Grouping data by 'Financial year' and sum the 'Total population' for 
# both BHC and AHC.
bhc_income_data_grouped = bhc_income_data.groupby('Financial year')\
    ['Total population'].sum()
ahc_income_data_grouped = ahc_income_data.groupby('Financial year')\
    ['Total population'].sum()

# Plotting the line chart.
plt.figure(figsize=(12, 6))
bhc_income_data_grouped.plot(
    kind='line', 
    marker='o', 
    color='teal', 
    label='Before Housing Cost - In low income'
)
ahc_income_data_grouped.plot(
    kind='line', 
    marker='o', 
    color='purple', 
    label='After Housing Cost - In low income'
)
plt.title('Total Population in Low Income by Financial Year (BHC vs AHC)')
plt.xlabel('Financial Year')
plt.ylabel('Total Population (Millions)')
plt.legend()
plt.grid(True)
plt.gca().yaxis.set_major_formatter\
    (plt.FuncFormatter(lambda x, _: f'{x/1e6:.1f}M'))
plt.show()
```

```python
# Exploring Housing Benefit.
benefit_analysis = df.pivot_table(
    index='Tenure type',
    columns='Housing benefit',
    values='Total population',
    aggfunc='sum'
)

# Plotting the stacked bar chart.
benefit_analysis.plot(kind='bar', stacked=True, figsize=(12,6), color=['teal', 'purple'])
plt.title('Housing Benefit Receipt by Tenure Type')
plt.ylabel('Total Population (Millions)')
plt.legend(title='Benefit Status', bbox_to_anchor=(1.05, 1))
plt.gca().yaxis.set_major_formatter\
    (plt.FuncFormatter(lambda x, _: f'{x/1e6:.1f}M'))
plt.show()
```

```python
# Exploring Regional Poverty Patterns.
regional_poverty_ahc = df.pivot_table(
    index='Region',
    columns='Median net household income(AHC)',
    values='Total population',
    aggfunc='sum'
).apply(lambda x: x/x.sum(), axis=1)

# Plotting the chart.
plt.figure(figsize=(12,6))
regional_poverty_ahc['In low income'].sort_values().\
    plot(kind='barh', color='teal')
plt.title('Proportion of Low-Income Households (AHC) by Region')
plt.xlabel('Percentage in Poverty')
plt.grid(axis='x')
plt.show()
```

```python
# Exploring Regional Poverty Patterns (BHC).
regional_poverty_bhc = df.pivot_table(
    index='Region',
    columns='Median net household income(BHC)',
    values='Total population',
    aggfunc='sum'
).apply(lambda x: x/x.sum(), axis=1)

# Plotting the chart.
plt.figure(figsize=(12,6))
regional_poverty_bhc['In low income'].sort_values().\
    plot(kind='barh', color='purple')
plt.title('Proportion of Low-Income Households (BHC) by Region')
plt.xlabel('Percentage in Poverty')
plt.grid(axis='x')
plt.show()
```

```python
# Combining AHC and BHC data for compare regional poverty before and
# after the housing cost.
combined_poverty = pd.DataFrame({
    'After Housing Cost': regional_poverty_ahc['In low income'],
    'Before Housing Cost': regional_poverty_bhc['In low income']
})

# Plotting the combined data.
combined_poverty.sort_values(by='After Housing Cost').plot(
    kind='barh', figsize=(12, 8), color=['purple', 'teal']
)
plt.title('Proportion of Low-Income Households by Region (AHC vs BHC)')
plt.xlabel('Percentage in Poverty')
plt.ylabel('Region')
plt.grid(axis='x')
plt.legend(title='Income Measure')
plt.tight_layout()
plt.show()
```

```python
# Exploring  which Tenure-Type push into poverty After Housing Cost.
tenure_poverty = df.pivot_table(
    index='Tenure type',
    columns='Median net household income(AHC)',
    values='Total population',
    aggfunc='sum'
).apply(lambda x: x/x.sum(), axis=1)

# Plotting the heatmap.
plt.figure(figsize=(10,6))
sns.heatmap(tenure_poverty, annot=True, fmt='.1%', cmap='YlOrRd')
plt.title('Poverty Distribution Across Tenure Types')
plt.ylabel('Tenure Type')
plt.xlabel('Income Status (After Housing Cost)')
plt.show()
```

```python
# Calculating poverty rates before/after housing costs by tenure type.
poverty_rates = df.pivot_table(
    index='Tenure type',
    columns=['Median net household income(BHC)', 'Median net household income(AHC)'],
    values='Total population',
    aggfunc='sum'
).fillna(0).apply(lambda x: x/x.sum()*100, axis=1)

# Simplifying column names and filtering key comparison.
poverty_rates.columns = ['_'.join(col).strip() for col in poverty_rates.\
                         columns.values]
poverty_rates = poverty_rates[['Not in low income_Not in low income', 
                             'Not in low income_In low income']]
poverty_rates.columns = ['Non-Poor (BHC→AHC)', 'Pushed into Poverty by Housing Costs']

# Visualising.
plt.figure(figsize=(12, 6))
ax = poverty_rates.plot(kind='bar', 
                       color=['#4c72b0', '#dd8452'],
                       edgecolor='black',
                       width=0.8,
                       figsize=(14, 7))
plt.title(
    'Impact of Housing Costs on Poverty Rates by Tenure Type', 
    pad=20, 
    fontsize=16
)
plt.xlabel('Tenure Type', fontsize=12, labelpad=15)
plt.ylabel('Percentage of Population', fontsize=12, labelpad=15)
plt.xticks(rotation=45, ha='right', fontsize=11)
plt.grid(axis='y', alpha=0.3)
plt.legend(title='Income Status', bbox_to_anchor=(1.05, 1))

# Adding percentage annotations.
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height:.1f}%', 
                (x + width/2, y + height + 1), 
                ha='center', 
                fontsize=10,
                color='black')

plt.text(0.5, -0.3, 
        "BHC = Before Housing Costs | AHC = After Housing Costs\n", 
        ha='center', 
        va='center', 
        transform=ax.transAxes,
        fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Housing Cost Impact.
poverty_gap = df.groupby(['Median net household income(BHC)', 'Median net household income(AHC)'])['Total population'].sum().unstack()
poverty_gap
```

```python
# Visualizing the poverty gap.
poverty_gap.plot(kind='bar', figsize=(10, 6), color=['#4c72b0', '#dd8452'])
plt.title('Poverty Gap Before and After Housing Costs')
plt.xlabel('Income Status Before Housing Costs')
plt.ylabel('Total Population')
plt.xticks(rotation=0)
plt.legend(title='Income Status After Housing Costs', bbox_to_anchor=(1.05, 1))
plt.grid(axis='y', alpha=0.3)
plt.gca().yaxis.set_major_formatter\
    (plt.FuncFormatter(lambda x, _: f'{x/1e6:.1f}M'))
# Adding population count annotations.
for p in plt.gca().patches:
    height = p.get_height()
    plt.gca().annotate(f'{height:.1f}', (p.get_x() + p.get_width() / 2., height),
                       ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.tight_layout()
plt.show()
```
