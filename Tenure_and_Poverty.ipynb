```python
# Exploring how Median net household income Before Housing Cost and 
# After Housing Cost are changing over the years for the In low income 
# gruop.

# Filtering the dataframe for 'In low income' in both 
# 'Median net household income(BHC)' and 'Median net household 
# income(AHC)'.
bhc_income_data = df[df['Median net household income(BHC)'] == 'In low income']
ahc_income_data = df[df['Median net household income(AHC)'] == 'In low income']

# Grouping data by 'Financial year' and sum the 'Total population' for 
# both BHC and AHC.
bhc_income_data_grouped = bhc_income_data.groupby('Financial year')\
    ['Total population'].sum()
ahc_income_data_grouped = ahc_income_data.groupby('Financial year')\
    ['Total population'].sum()

# Plotting the line chart.
plt.figure(figsize=(12, 6))
bhc_income_data_grouped.plot(
    kind='line', 
    marker='o', 
    color='teal', 
    label='Before Housing Cost - In low income'
)
ahc_income_data_grouped.plot(
    kind='line', 
    marker='o', 
    color='purple', 
    label='After Housing Cost - In low income'
)
plt.title('Total Population in Low Income by Financial Year (BHC vs AHC)')
plt.xlabel('Financial Year')
plt.ylabel('Total Population (Millions)')
plt.legend()
plt.grid(True)
plt.gca().yaxis.set_major_formatter\
    (plt.FuncFormatter(lambda x, _: f'{x/1e6:.1f}M'))
plt.show()
```

**Key Insights**
1. Low income levels are consistently higher after housing costs (AHC) compared to before housing costs (BHC), indicating that housing costs significantly impact the financial well-being of a substantial portion of the population.

```python
# Exploring Regional Poverty Patterns in Financial year 2022-23.
regional_poverty_ahc = df_recent_year.pivot_table(
    index='Region',
    columns='Median net household income(AHC)',
    values='Total population',
    aggfunc='sum'
).apply(lambda x: x/x.sum(), axis=1)

# Plotting the chart.
plt.figure(figsize=(12,6))
regional_poverty_ahc['In low income'].sort_values().\
    plot(kind='barh', color='teal')
plt.title('Proportion of Low-Income Households (AHC) by Region - (2022-23)')
plt.xlabel('Percentage in Poverty')
plt.grid(axis='x')
plt.show()
```

**Key Insights**  
1. London has the highest proportion of low-income households after housing costs (AHC), likely due to its high cost of living and expensive housing market.  
2. Regions like the East and East Midlands show lower proportions of low-income households, suggesting better affordability or higher income levels relative to housing costs.  
3. Southern regions also exhibit significant low-income rates, indicating regional economic disparities and potential challenges in housing affordability and income levels.  
4. The data highlights regional inequalities, emphasizing the need for targeted policies to address poverty and housing affordability in the most affected areas.

```python
# Exploring Regional Poverty Patterns (BHC) in Financial year 2022-23.
regional_poverty_bhc = df_recent_year.pivot_table(
    index='Region',
    columns='Median net household income(BHC)',
    values='Total population',
    aggfunc='sum'
).apply(lambda x: x/x.sum(), axis=1)

# Plotting the chart.
plt.figure(figsize=(12,6))
regional_poverty_bhc['In low income'].sort_values().\
    plot(kind='barh', color='purple')
plt.title('Proportion of Low-Income Households (BHC) by Region - (2022-23)')
plt.xlabel('Percentage in Poverty')
plt.grid(axis='x')
plt.show()
```

```python
# Combining AHC and BHC data for compare regional poverty before and
# after the housing cost in Finanacial year 2022-23.
combined_poverty = pd.DataFrame({
    'After Housing Cost': regional_poverty_ahc['In low income'],
    'Before Housing Cost': regional_poverty_bhc['In low income']
})

# Plotting the combined data.
combined_poverty.sort_values(by='After Housing Cost').plot(
    kind='barh', figsize=(12, 8), color=['purple', 'teal']
)
plt.title('Proportion of Low-Income Households by Region (AHC vs BHC) - (2022-23)')
plt.xlabel('Percentage in Poverty')
plt.ylabel('Region')
plt.grid(axis='x')
plt.legend(title='Income Measure')
plt.tight_layout()
plt.show()
```

**Key Insights**  
1. Low-income proportions are consistently higher after housing costs (AHC) across all regions, highlighting the significant impact of housing expenses on household finances.

2. London shows the largest gap between AHC and BHC low-income rates, underscoring the severe effect of high housing costs in the capital.

3. Northern regions exhibit notable increases in low-income rates after housing costs, indicating housing affordability challenges in these areas.

4. East regions like the East and East Midlands have relatively lower low-income rates, suggesting better income-to-housing cost ratios compared to other regions.
  
5. The data emphasizes regional disparities, pointing to the need for region-specific policies to address housing affordability and income support.

```python
# Exploring  which Tenure-Type push into poverty After Housing Cost in 
# Financial year 2022-23.
tenure_poverty = df_recent_year.pivot_table(
    index='Tenure type',
    columns='Median net household income(AHC)',
    values='Total population',
    aggfunc='sum'
).apply(lambda x: x/x.sum(), axis=1)

# Plotting the heatmap.
plt.figure(figsize=(10,6))
sns.heatmap(tenure_poverty, annot=True, fmt='.1%', cmap='YlOrRd')
plt.title('Poverty Distribution Across Tenure Types - (2022-23)')
plt.ylabel('Tenure Type')
plt.xlabel('Income Status (After Housing Cost)')
plt.show()
```

**Key Insights**  
1. Social rented sector tenants have the highest poverty rate (44%), reflecting the economic challenges faced by individuals relying on social housing.

2. Privately rented households also show a significant poverty rate (35.2%), likely due to high rental costs and financial instability in the private rental market.

3. Homeowners, particularly those with mortgages, have the lowest poverty rates (9.5%), indicating greater financial stability and lower housing cost burdens compared to renters.

4. Outright homeowners have a moderate poverty rate (14.3%), suggesting that while owning a home outright provides financial security, some still face income challenges.
  
5. The data highlights the correlation between tenure type and poverty, emphasizing the need for targeted support to address financial instability among renters and social housing tenants.

```python
# Calculating poverty rates before/after housing costs by tenure type in
# Fianancial year 2022-23.
poverty_rates = df_recent_year.pivot_table(
    index='Tenure type',
    columns=['Median net household income(BHC)', 'Median net household income(AHC)'],
    values='Total population',
    aggfunc='sum'
).fillna(0).apply(lambda x: x/x.sum()*100, axis=1)

# Simplifying column names and filtering key comparison.
poverty_rates.columns = ['_'.join(col).strip() for col in poverty_rates.\
                         columns.values]
poverty_rates = poverty_rates[['Not in low income_Not in low income', 
                             'Not in low income_In low income']]
poverty_rates.columns = ['Non-Poor (BHC→AHC)', 'Pushed into Poverty by Housing Costs']

# Visualising.
plt.figure(figsize=(12, 6))
ax = poverty_rates.plot(kind='bar', 
                       color=['#4c72b0', '#dd8452'],
                       edgecolor='black',
                       width=0.8,
                       figsize=(14, 7))
plt.title(
    'Impact of Housing Costs on Poverty Rates by Tenure Type - (2022-23)', 
    pad=20, 
    fontsize=16
)
plt.xlabel('Tenure Type', fontsize=12, labelpad=15)
plt.ylabel('Percentage of Population', fontsize=12, labelpad=15)
plt.xticks(rotation=45, ha='right', fontsize=11)
plt.grid(axis='y', alpha=0.3)
plt.legend(title='Income Status', bbox_to_anchor=(1.05, 1))

# Adding percentage annotations.
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height:.1f}%', 
                (x + width/2, y + height + 1), 
                ha='center', 
                fontsize=10,
                color='black')

plt.text(0.5, -0.3, 
        "BHC = Before Housing Costs | AHC = After Housing Costs\n", 
        ha='center', 
        va='center', 
        transform=ax.transAxes,
        fontsize=10)

plt.tight_layout()
plt.show()
```

**Key Insights**
1. Housing costs significantly impact poverty rates, with a substantial portion of the population affected across different tenure types.

2. The highest impact is observed in rented sectors, where a large percentage of tenants face financial strain due to housing expenses.

3. Homeowners, particularly those with mortgages, experience lower poverty rates (1.4%), indicating that housing costs are more manageable for this group.

4. Outright homeowners have the lowest poverty rates, suggesting that owning a home without ongoing mortgage payments provides greater financial stability.

```python
# Housing Cost Impact.
poverty_gap = df_recent_year.groupby(['Median net household income(BHC)', 'Median net household income(AHC)'])['Total population'].sum().unstack()
poverty_gap
```

```python
# Visualizing the poverty gap.
poverty_gap.plot(kind='bar', figsize=(10, 6), color=['#4c72b0', '#dd8452'])
plt.title('Poverty Gap Before and After Housing Costs')
plt.xlabel('Income Status Before Housing Costs')
plt.ylabel('Total Population')
plt.xticks(rotation=0)
plt.legend(title='Income Status After Housing Costs', bbox_to_anchor=(1.05, 1))
plt.grid(axis='y', alpha=0.3)
plt.gca().yaxis.set_major_formatter\
    (plt.FuncFormatter(lambda x, _: f'{x/1e6:.1f}M'))
# Adding population count annotations.
for p in plt.gca().patches:
    height = p.get_height()
    plt.gca().annotate(f'{height:.1f}', (p.get_x() + p.get_width() / 2., height),
                       ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.tight_layout()
plt.show()
```

**Key Insights**
1. A significant portion of the population falls into low income after housing costs, highlighting the substantial impact of housing expenses on financial stability.

2. The poverty gap widens after accounting for housing costs, indicating that housing affordability is a critical factor in determining overall poverty levels.

3. The data suggests that housing costs push a considerable number of households into low income, emphasizing the need for policies aimed at reducing housing expenses and providing financial support to vulnerable populations.
